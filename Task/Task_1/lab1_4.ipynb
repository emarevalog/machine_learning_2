```python
# Cargar la imagen de tu rostro
image_path2 = r"C:\Users\DO IT WISER\Documents\Personal\machine_learning2\Task\Task_1\img_personal\Melissa_Arevalo_1_resized.jpeg"  # Reemplaza con la ruta de tu imagen
image2 = Image.open(image_path2)

# Convertir la imagen a escala de grises si es necesario
image_gray2 = image2.convert('L')

# Lista de valores singulares a probar
singular_values = [1, 10, 20, 50, 100]

# Crear una figura para mostrar las imágenes
plt.figure(figsize=(15, 8))

# Mostrar la imagen original
plt.subplot(2, 3, 1)
plt.imshow(image_gray2, cmap='gray')
plt.title('Imagen original')
plt.axis('off')

# Iterar sobre diferentes cantidades de valores singulares
for i, num_singular_values in enumerate(singular_values):
    # Aplicar SVD con el número correspondiente de valores singulares
    svd = SVD(n_components=num_singular_values)
    reconstructed_img = svd.fit_transform(np.array(image_gray2))
    
    # Mostrar la imagen reconstruida
    plt.subplot(2, 3, i+2)
    plt.imshow(reconstructed_img, cmap='gray')
    plt.title(f'Con {num_singular_values} valores singulares')
    plt.axis('off')

plt.tight_layout()
plt.show()

```
